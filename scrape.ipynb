```python
base_url = "https://en.wikipedia.org/wiki/{}"
output_directory = "wikipedia_data/"
for topic in topics:
    formatted_topic = topic.replace(" ", "_")
    url = base_url.format(formatted_topic)
    try:
        response = requests.get(url)
        response.raise_for_status()  # Raise an HTTPError for bad responses

        # Parse the HTML content with BeautifulSoup
        soup = BeautifulSoup(response.text, "html.parser")

        # Find the element with id "bodyContent"
        body_content = soup.find("div", {"id": "bodyContent"})

        if body_content:
            # Extract text from the bodyContent element
            content_text = body_content.get_text()

            # Save the text to a file
            file_path = f"{output_directory}{formatted_topic}.txt"
            with open(file_path, "w", encoding="utf-8") as file:
                file.write(content_text)

            print(
                f"Successfully retrieved and saved content for {formatted_topic} in {file_path}"
            )
        else:
            print(f"Error: Could not find 'bodyContent' for {formatted_topic}")

    except requests.exceptions.RequestException as e:
        print(f"Error fetching content for {url}: {e}")
        pass  # Continue to the next iteration even if there's an exception
```

```python
import os
output_file_path = "wikipedia_data/final.txt"  # Specify the path for the final text file
try:
    with open(output_file_path, 'w', encoding='utf-8') as final_file:
        for topic in topics:
            formatted_topic = topic.replace(" ", "_")
            file_path = f"{output_directory}{formatted_topic}.txt"

            if os.path.exists(file_path):
                try:
                    with open(file_path, 'r', encoding='utf-8') as topic_file:
                        # Read the content from the topic file
                        content_text = topic_file.read()

                        # Append the content to the final file
                        final_file.write(f"\n\n==== {formatted_topic} ====\n\n")
                        final_file.write(content_text)

                        print(f"Successfully read and appended content for {formatted_topic}")
                except IOError as e:
                    print(f"Error reading content from {file_path}: {e}")
            else:
                print(f"Error: File not found for {formatted_topic}")

except IOError as e:
    print(f"Error writing to the final file: {e}")

print("Finished processing topics.")
```
